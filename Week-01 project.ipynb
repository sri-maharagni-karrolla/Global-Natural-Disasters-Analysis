Project Title :
Analysis of Global Natural Disasters (1900–2021) for Climate Risk and Disaster Management

Problem Statement : 
Natural disasters such as floods, earthquakes, and storms have been increasing in frequency and severity due to climate change. Understanding past disaster patterns is essential for assessing risks and supporting disaster management strategies.

Project Description : 
This project uses the Global Natural Disasters Dataset (1900–2021) to perform basic data understanding. The analysis includes dataset structure, statistical summaries, and missing value detection, laying the groundwork for deeper insights in upcoming week

```python
# Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Display all columns in outputs
pd.set_option('display.max_columns', None)
```

```python
# Load CSV files
df1 = pd.read_csv(r"D:\Documents\Global natural disaster\DISASTERS\1900_2021_DISASTERS.xlsx - emdat data.csv")
df2 = pd.read_csv(r"D:\Documents\Global natural disaster\DISASTERS\1970-2021_DISASTERS.xlsx - emdat data.csv")
```

```python
# Check first few rows
print(df1.head())
print(df2.head())
```

```python
# Check shape (rows, columns)
print(df1.shape)
print(df2.shape)
```

```python
# Check column names
print(df1.columns)
print(df2.columns)
```

```python
# Dataset info
print(df1.info())
print(df2.info())
```

```python
# Statistical summary
print(df1.describe())
print(df2.describe())
```

```python
# Missing values
print(df1.isnull().sum())
print(df2.isnull().sum())
```

```python
# Combine both datasets
df = pd.concat([df1, df2], ignore_index=True)
print(df.shape)
```

```python
# Count of disasters by year
plt.figure(figsize=(12,6))
sns.countplot(x='Year', data=df, order=sorted(df['Year'].unique()))
plt.xticks(rotation=90)
plt.title("Number of Disasters per Year")
plt.show()
```

```python
# Top disaster types
plt.figure(figsize=(10,6))
df['Disaster Type'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Disaster Types")
plt.ylabel("Count")
plt.show()
```
